## Making the preprocessing pipeline

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd

# Define columns
selected_features = ["Age", "Pclass", "SibSp", "Parch", "Fare", "Sex", "Embarked"]
target = "Survived"

# Split numerical columns into two groups
log_transform_cols = ["Fare", "SibSp", "Parch"]
regular_num_cols = ["Age", "Pclass"]
categorical_cols = ["Sex", "Embarked"]

# Create log transformer
log_transformer = FunctionTransformer(lambda x: np.log(x + 1e-9), inverse_func=np.exp)

# Categorical transformer remains the same
class CategoricalTransformer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        X["Sex"] = X["Sex"].map({"male": 0, "female": 1})
        X = pd.get_dummies(X, columns=["Embarked"], drop_first=True)
        return X

# Create two different numerical preprocessors
log_num_preprocessor = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("log_transform", log_transformer),
    ("scaler", StandardScaler()),
])

regular_num_preprocessor = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])

# Combine all preprocessors in ColumnTransformer
preprocessor = ColumnTransformer([
    ("log_num", log_num_preprocessor, log_transform_cols),
    ("regular_num", regular_num_preprocessor, regular_num_cols),
    ("cat", CategoricalTransformer(), categorical_cols),
])

# Create final pipeline
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(n_estimators=500, random_state=42)),
])
```
